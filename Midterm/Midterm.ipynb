>It appears there exists duplicate data in the raw dataset, we need to find a way to drop the dirty duplicated data.

```python
LoanID_counts = df['Loan_ID'].value_counts()
print('Number of distinct Loan_IDs is', len(LoanID_counts.index))
repeated_LoanIDs = LoanID_counts[LoanID_counts > 1].index
unique_LoanIDs = LoanID_counts[LoanID_counts == 1].index
```

```python
unique_records = df[df['Loan_ID'].isin(unique_LoanIDs)]
repeated_records = df[df['Loan_ID'].isin(repeated_LoanIDs)]

def merge_repeated_records(group):
    combined_record={}
    for col in group.columns:
        if (group[col].dtype=='float64'):
            combined_record[col] = group[col].dropna().median()
        elif (group[col].dtype=='int64'):
            l = group[col].values.tolist()
            combined_record[col] = int(max(set(l),key=l.count))
        else:
            l = group[col].values.tolist()
            combined_record[col] = max(set(l),key=l.count)
    return combined_record


cleaned_records = pd.DataFrame(repeated_records.groupby('Loan_ID').apply(merge_repeated_records).values.tolist())

df = pd.concat([unique_records, cleaned_records], ignore_index = True)[df.columns]
```

### 2. Use logistic regression to analyze this data.

<b>(a) Use 100 bootstrap samples to estimate the the standard errors of the coefficients from the logistic regression fit AUC. Provide a 68% and 95% confidence interval for the model parameters of the logistic regression.</b>

```python
predictors = list(train_df.columns)
predictors.remove('Loan_ID')
predictors.remove('Customer_ID')
predictors.remove('Loan_Status')
predictors.remove('Term')
predictors.remove('Years_in_current_job')
predictors.remove('Home_Ownership')
predictors.remove('Purpose')
predictors.remove('Monthly_Debt')
predictors
```

```python
sample_df = train_df.sample(n=100)
Y_train = np.array(sample_df['Loan_Status'])
X_train = np.array(sample_df[predictors])
Y_train[Y_train == 'Fully Paid'] = 1
Y_train[Y_train == 'Charged Off'] = 0
Y_train = Y_train.astype('int')

X_test = np.array(test_df[predictors])
Y_test = np.array(test_df['Loan_Status'])
Y_test[Y_test == 'Fully Paid'] = 1
Y_test[Y_test == 'Charged Off'] = 0
Y_test = Y_test.astype('int')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.metrics import roc_curve, auc
```

```python
logreg_clf = LogisticRegression()
# Fit model. let X_train = matric of predictors, Y_train = matric of variables
res_logreg_clf = logreg_clf.fit(X_train, Y_train)

# Predict values fro training set
Y_pred = res_logreg_clf.predict(X_test)

# Confusion matrix
confusion = confusion_matrix(Y_test, Y_pred)
TP = confusion[1, 1]
TN = confusion[0, 0]
FP = confusion[0, 1]
FN = confusion[1, 0]
print(TP, TN, FP, FN)
```

```python
# Metrics computed from a confusion matrix
# (TP + TN) / float(TP + TN + FP + FN)
print('Accuracy score:', accuracy_score(Y_test, Y_pred))

# Calculate TP rate or recall
# TP / float(TP + FN)
print('Recall score:', recall_score(Y_test, Y_pred))

# Precision: When a positive value is predicted, how often is the prediction correct?
# How "precise" is the classifier when predicting positive instances?
# TP / float(TP + FP)
print('Precision score:', precision_score(Y_test, Y_pred))
```

```python
fpr, tpr, thresholds = roc_curve(Y_test, Y_pred, pos_label=1)
print('fpr:', fpr)
print('tpr:', tpr)
print('thresholds:', thresholds)
```

```python
plt.plot(fpr, tpr)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.title('ROC curve for loan status classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.grid(True)
plt.show()
```

<b>(d) Use 5-fold cross-validation and Forward Stepwise Selection approach for feature selection. In each iteration, choose the best model using the highest AUC. Report the features and AUC that appears to provide the best results.</b>

```python
def caculate_auc(features):
    X_train = np.array(train_df[features])
    Y_train = np.array(train_df['Loan_Status'])
    Y_train[Y_train == 'Fully Paid'] = 1
    Y_train[Y_train == 'Charged Off'] = 0
    Y_train = Y_train.astype('int')
    
    logit = LogisticRegression()
    res_logit = logit.fit(X_train, Y_train)
    
    X_test = np.array(test_df[features])
    Y_test = np.array(test_df['Loan_Status'])
    Y_test[Y_test == 'Fully Paid'] = 1
    Y_test[Y_test == 'Charged Off'] = 0
    Y_test = Y_test.astype('int')
    
    return cross_val_score(res_logit, X_test, Y_test, cv=5, scoring='roc_auc').mean()
```

```python
# first round iteration
for p in predictors:
    auc_res = caculate_auc([p])
    print([p], auc_res)
```
